```python
from tkinter import ttk
import tkinter as tk
def showRoomTime(key , batch_dict,a):
    print(key)
    screen3 = tk.Tk()
    screen3.geometry("1000x800")
    screen3.title("BATCH 2K24")
    
    canvas3 = tk.Canvas(screen3)
    canvas3.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

    scrollbar3 = ttk.Scrollbar(screen3, orient="vertical", command=canvas3.yview)
    scrollbar3.pack(side=tk.RIGHT, fill="y")
    canvas3.configure(yscrollcommand=scrollbar3.set)

    scrollable_frame3 = ttk.Frame(canvas3)

    canvas3.create_window((0, 0), window=scrollable_frame3, anchor="nw")

    scrollable_frame3.bind("<Configure>", lambda e: canvas3.configure(scrollregion=canvas3.bbox("all")))

    canvas3.bind_all("<MouseWheel>", lambda event:canvas3.yview_scroll(int(-1 * (event.delta / 120)), "units") )  
    table = ttk.Treeview(scrollable_frame3,columns=("Room" , "Time"),show="headings")
    table.pack(expand=True,fill=tk.BOTH)
    
    style = ttk.Style()
    style.configure("Treeview.Heading",font = ('Arial' , 20 , "bold"))
    
    
    table.heading("Room" , text="Room" )
    table.heading("Time" , text="Time")
    table.column("Room",width=500,anchor="center")
    table.column("Time",width=500,anchor="center")
    for i,j in batch_dict.items():
        if i == a:
            for k,l in j.items():
                if(k == key):
                    for a,b in zip(l["Room"],l["Time"]) :
                        table.insert("",tk.END,values=(a,b))

    screen3.mainloop()    
        
```
